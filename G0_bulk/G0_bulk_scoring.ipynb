```python
# Filter the genes for those that are in tumour_adata.var_names
upregulated_genes = [gene for gene in upregulated_genes if gene in mat_expr_scoring.columns.tolist()]
downregulated_genes = [gene for gene in downregulated_genes if gene in mat_expr_scoring.columns.tolist()]
```

```python
gene_expr = mat_expr_scoring.T
```

```python
gene_expr
```

```python
gene_expr = pd.DataFrame(gene_expr)
```

```python
from scipy.stats import zscore

gene_expr_score = zscore(gene_expr, axis=1)
```

```python
gene_expr_score
```

```python
gene_expr_score = pd.DataFrame(gene_expr_score)
```

```python
gene_expr_score
```

```python
gene_expr_score.index = gene_expr.index
```

```python
# Calculate the scores for tumour cells
pos_score = gene_expr_score.loc[upregulated_genes].sum(axis=0) / np.sqrt(len(upregulated_genes))
neg_score = gene_expr_score.loc[downregulated_genes].sum(axis=0) / np.sqrt(len(downregulated_genes))

# Compute the z-score differences for tumour cells
zscore_differences = neg_score - pos_score
```

```python
zscore_differences.to_csv("BULK_PROLIF_SCORE.csv", sep=",", index=True, header=True)
```
